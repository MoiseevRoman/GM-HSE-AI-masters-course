```python
import requests
response = requests.get('http://storage.mds.yandex.net:80/get-voicetoloka/2021744/37e917e1-1fdf-4064-a74f-57a84bcb28b9')
wav = response.content
```

```python
import pandas as pd

df = pd.read_csv('urls_normalized.tsv', sep='\t', header=None)
df.columns=['links']
df
```

```python
from transformers import WhisperForConditionalGeneration, WhisperProcessor
import torch
import torchaudio
from tqdm import tqdm
import requests
device = "cuda:0" if torch.cuda.is_available() else "cpu"


processor = WhisperProcessor.from_pretrained("openai/whisper-small")
model = WhisperForConditionalGeneration.from_pretrained("openai/whisper-small").to(device)
forced_decoder_ids = processor.get_decoder_prompt_ids(language="russian", task="transcribe")

wavs = []
for row in tqdm(df.values):
  resp = requests.get(row[0]).content
  wav, sr = torchaudio.load(resp)

  wav = wav.mean(dim=0)

  # пересэмплируйте имеющиеся аудио в 16 kHz - пример кода ниже, можете его менять
  if sr != 16000:
      wav = torchaudio.functional.resample(wav, orig_freq=sr, new_freq=16000)
      sr = 16000
  input_features = processor(wav, sampling_rate=16000, return_tensors="pt").input_features.to(device)
  predicted_ids = model.generate(input_features, forced_decoder_ids=forced_decoder_ids)
  transcription = processor.batch_decode(predicted_ids, skip_special_tokens=True)
  wavs.append(transcription)
print(wavs[0])
# ваш код здесь

# также вам может пригодиться установить распознавание русского языка как гиперпараметр whisper-small, так как наш датасет русскоязычный
```

```python
df['texts'] = wavs
df['texts'] = df['texts'].apply(lambda x: x[0])
test_df = df.sample(10)
test_df
```

## Шаг 2 (1 балл)

Текст распознается с ошибками.
Попробуйте исправить ошибки с помощью готовой (предобученной) модели spell correction.

Выведите на экран 10 текстов с предыдущего шага и их исправления с помощью модели https://huggingface.co/UrukHan/t5-russian-spell.

```python
from transformers import AutoModelForSeq2SeqLM, T5TokenizerFast

# Зададим название выбронной модели из хаба
MODEL_NAME = 'UrukHan/t5-russian-spell'
MAX_INPUT = 256

# Загрузка модели и токенизатора
tokenizer = T5TokenizerFast.from_pretrained(MODEL_NAME)
model = AutoModelForSeq2SeqLM.from_pretrained(MODEL_NAME)

# Входные данные (можно массив фраз или текст)
input_sequences = test_df['texts'].tolist()

task_prefix = "Spell correct: "                 # Токенизирование данных
encoded = tokenizer(
  [task_prefix + sequence for sequence in input_sequences],
  padding="longest",
  max_length=MAX_INPUT,
  truncation=True,
  return_tensors="pt",
)

predicts = model.generate(encoded.input_ids)    # # Прогнозирование

test_df['cor_texts'] = tokenizer.batch_decode(predicts, skip_special_tokens=True)  # Декодируем данные

```
